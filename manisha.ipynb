```python
source_school_df = pd.read_csv('2015_berkeley.csv', encoding='utf-16')
# print source_school_df[[u' Measure Names']].values

admits_source_school = source_school_df[source_school_df[u' Measure Names'].isin([u' Enrollees'])]

def floordiv(x, base=3):
    return math.ceil(x / float(base)) * base # rounds up to nearest multiple of three
    
admits_source_school[[u' Measure Values']] = admits_source_school[[u' Measure Values']].replace(' ', 1)
admits_source_school[[u' Measure Values']] = admits_source_school[[u' Measure Values']].astype(int)
admits_source_school['Enrollees'] = admits_source_school[[u' Measure Values']].apply(floordiv, axis=1)

print admits_source_school
```
